```python
import matplotlib.pyplot as plt
import random
```

```python
# Generate non-linear data
def generate_data():
    data = []
    for i in range(100):
        x = random.uniform(0, 10)
        y = x**2 + random.uniform(-10, 10)  # Quadratic relationship with some noise
        data.append((x, y))
    return data
```

```python
# Split data into training and testing sets
def split_data(data, split_ratio=0.8):
    random.shuffle(data)
    train_size = int(len(data) * split_ratio)
    train_data = data[:train_size]
    test_data = data[train_size:]
    return train_data, test_data
```

```python
# Calculate coefficients (intercept and slope) of the regression line
def coefficients(train_data):
    x_values = [x for x, _ in train_data]
    y_values = [y for _, y in train_data]
    x_mean = sum(x_values) / len(x_values)
    y_mean = sum(y_values) / len(y_values)
    numerator = sum((x - x_mean) * (y - y_mean) for x, y in train_data)
    denominator = sum((x - x_mean)**2 for x, _ in train_data)
    b1 = numerator / denominator  # Slope (b1) formula
    b0 = y_mean - b1 * x_mean  # Intercept (b0) formula
    return b0, b1
```

```python
# Make predictions for test data
def predict(test_data, b0, b1):
    return [b0 + b1 * x for x, _ in test_data]
```

```python
# Generate data
data = generate_data()
```

```python
# Split data into training and testing sets
train_data, test_data = split_data(data)
```

```python
# Calculate coefficients
b0, b1 = coefficients(train_data)
```

```python
# Make predictions for test data
predictions = predict(test_data, b0, b1)
```

```python
# Plotting
plt.scatter([x for x, _ in train_data], [y for _, y in train_data], color='blue', label='Training Data')  # Plot training data
plt.scatter([x for x, _ in test_data], [y for _, y in test_data], color='green', label='Testing Data')  # Plot testing data
plt.plot([x for x, _ in test_data], predictions, color='red', label='Regression Line')  # Plot regression line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with Training and Testing Data')
plt.legend()
plt.grid(True)
plt.show()
```
